```python
# 2) Write a Python program to compute the variance and standard deviation of a dataset

import math

def calculate_mean(data):
    """Calculates the mean of a list of numbers."""
    return sum(data) / len(data)

def calculate_variance(data, mean):
  """Calculates the variance of a list of numbers."""
  return sum((x - mean) ** 2 for x in data) / (len(data) - 1)

def calculate_std_dev(variance):
    """Calculates the standard deviation from the variance."""
    return math.sqrt(variance)

# Example usage:
data = [10, 12, 15, 18, 20]

# Calculate the mean
mean = calculate_mean(data)

# Calculate the variance
variance = calculate_variance(data, mean)

# Calculate the standard deviation
std_dev = calculate_std_dev(variance)

# Print the results
print("Data:", data)
print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# 9)  write a Python script to calculate covariance between two datasets
import numpy as np

def calculate_covariance(data1, data2):
    """
    Calculate the covariance between two datasets.

    Args:
        data1 (list or numpy.ndarray): The first dataset.
        data2 (list or numpy.ndarray): The second dataset.

    Returns:
        float: The covariance between the two datasets.

    Raises:
        ValueError: If the datasets have different lengths.
    """

    if len(data1) != len(data2):
        raise ValueError("Datasets must have the same length")

    return np.cov(data1, data2)[0, 1]

if __name__ == '__main__':
    dataset1 = [2, 4, 6, 8, 10]
    dataset2 = [1, 3, 5, 7, 9]

    try:
        covariance = calculate_covariance(dataset1, dataset2)
        print(f"Covariance: {covariance}")
    except ValueError as e:
        print(f"Error: {e}")

    dataset3 = [1, 2, 3]
    dataset4 = [4, 5]

    try:
       covariance = calculate_covariance(dataset3, dataset4)
       print(f"Covariance: {covariance}")
    except ValueError as e:
        print(f"Error: {e}")
```

```python
# 10 Write a Python script to calculate the correlation coefficient between two datasets
import numpy as np

def calculate_correlation_coefficient(dataset1, dataset2):
    """
    Calculates the Pearson correlation coefficient between two datasets.

    Args:
        dataset1 (list or numpy.ndarray): The first dataset.
        dataset2 (list or numpy.ndarray): The second dataset.

    Returns:
        float: The Pearson correlation coefficient, or None if an error occurs.
    """
    try:
        correlation_coefficient = np.corrcoef(dataset1, dataset2)[0, 1]
        return correlation_coefficient
    except Exception as e:
         print(f"An error occurred: {e}")
         return None

if __name__ == "__main__":
    data1 = [1, 2, 3, 4, 5]
    data2 = [2, 4, 5, 4, 5]

    correlation = calculate_correlation_coefficient(data1, data2)

    if correlation is not None:
        print(f"The Pearson correlation coefficient is: {correlation}")
```

```python
# 22) Generate a dataset and implement both variance and standard deviation computations
import math

data = [1, 2, 3, 4, 5]

# Calculate the mean
mean = sum(data) / len(data)

# Calculate the variance (sample variance)
squared_differences = [(x - mean) ** 2 for x in data]
variance = sum(squared_differences) / (len(data) - 1)

# Calculate the standard deviation
standard_deviation = math.sqrt(variance)

print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {standard_deviation}")
```

```python
# 23) Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

# Generate sample data with varying skewness and kurtosis
np.random.seed(0)
data_normal = np.random.normal(0, 1, 1000)
data_skewed_pos = np.random.exponential(1, 1000)
data_skewed_neg = -np.random.exponential(1, 1000)
data_kurtosis_high = np.random.laplace(0, 1, 1000)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot normal distribution
sns.histplot(data_normal, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title(f'Normal Distribution\nSkewness: {skew(data_normal):.2f}, Kurtosis: {kurtosis(data_normal):.2f}')

# Plot positively skewed distribution
sns.histplot(data_skewed_pos, kde=True, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title(f'Positively Skewed Distribution\nSkewness: {skew(data_skewed_pos):.2f}, Kurtosis: {kurtosis(data_skewed_pos):.2f}')

# Plot negatively skewed distribution
sns.histplot(data_skewed_neg, kde=True, ax=axes[1, 0], color='salmon')
axes[1, 0].set_title(f'Negatively Skewed Distribution\nSkewness: {skew(data_skewed_neg):.2f}, Kurtosis: {kurtosis(data_skewed_neg):.2f}')

# Plot high kurtosis distribution
sns.histplot(data_kurtosis_high, kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title(f'High Kurtosis Distribution\nSkewness: {skew(data_kurtosis_high):.2f}, Kurtosis: {kurtosis(data_kurtosis_high):.2f}')

plt.tight_layout()
plt.show()
```
